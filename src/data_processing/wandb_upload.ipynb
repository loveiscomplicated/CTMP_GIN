```python
import wandb
import json
import os

def upload_jsonl_to_wandb(file_path, project_name, run_name=None, config=None):
    """
    JSONL 파일의 로그 데이터를 WandB에 업로드합니다.
    
    Args:
        file_path (str): .jsonl 파일 경로
        project_name (str): WandB 프로젝트 이름
        run_name (str, optional): 실행(Run) 이름. 미지정 시 자동 생성.
        config (dict, optional): 모델 설정값 등 실험 메타데이터
    """
    if not os.path.exists(file_path):
        print(f"Error: 파일을 찾을 수 없습니다. ({file_path})")
        return

    # 1. WandB 초기화
    wandb.init(project=project_name, name=run_name, config=config)

    print(f"Uploading {file_path} to WandB project '{project_name}'...")

    # 2. 파일 읽기 및 로깅
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            
            # 'epoch'를 step으로 사용
            if "epoch" in data:
                step = data.pop("epoch")
                wandb.log(data, step=step)
            else:
                # epoch 정보가 없는 라인(예: 최종 test 결과)은 그냥 로깅
                wandb.log(data)

    # 3. 종료
    wandb.finish()
    print("Upload complete!")
```

```python
import os
cur_dir = os.getcwd()
runs_protected_path = os.path.join(cur_dir, '..', '..', 'runs', 'protected')
ctmp_gin_1 = os.path.join(runs_protected_path, 'three_seeds', '20260301-205342__ctmp_gin__bs=256__lr=2.00e-04__seed=1', 'metrics.jsonl')
ctmp_gin_2 = os.path.join(runs_protected_path, 'three_seeds', '20260301-205352__ctmp_gin__bs=256__lr=2.00e-04__seed=2', 'metrics.jsonl')
ctmp_gin_3 = os.path.join(runs_protected_path, 'three_seeds', '20260301-205416__ctmp_gin__bs=256__lr=2.00e-04__seed=3', 'metrics.jsonl')

upload_jsonl_to_wandb(ctmp_gin_1, "ctmp_gin", run_name="ctmp_gin_opt_seed_1")
upload_jsonl_to_wandb(ctmp_gin_2, "ctmp_gin", run_name="ctmp_gin_opt_seed_2")
upload_jsonl_to_wandb(ctmp_gin_3, "ctmp_gin", run_name="ctmp_gin_opt_seed_3")
```
